```python
# import the libraries
import os
import google.generativeai as genai
from langchain.vectorstores import FAISS  # this will be used to create the vector database
from langchain_community.embeddings import HuggingFaceEmbeddings # to perform word embeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter # to split the text into chunks
from pypdf import PdfReader # to read the pdf file
import faiss  # to perform similarity search
```

```python
# step 1: Configure the models

# Configure LLM
key = os.getenv("GOOGLE_API_KEY")
genai.configure(api_key=key)
llm_model=genai.GenerativeModel('gemini-2.5-flash-lite')

#Configure the  Embedding model
embedding_model = HuggingFaceEmbeddings(model_name="all-miniLM-L6-v2")
```

```python
#step 2: Loading the pdf file and extract the text.
loaded_file=PdfReader("Rag Chatbot.pdf")

raw_text=" "
for page in loaded_file.pages:
    text_only=page.extract_text()
    if text_only:
        raw_text+=text_only
```

```python
print(raw_text)
```

```python
# Step 3: Chunking (create Chunks)
splitter=RecursiveCharacterTextSplitter(chunk_size=300,chunk_overlap=50)
chunks=splitter.split_text(raw_text)
```

```python
len(chunks)
```

```python
print(chunks[0])
```

```python
# Step 4: Create FAISS vector store
vectorstore=FAISS.from_texts(chunks,embedding_model)
```

```python
# Step 5: Configure retriever
retriever=vectorstore.as_retriever(search_kwargs={"k":5})
```

```python
# Step 6: Take the query
query="Show me the steps to proceed with this project."
```

```python
# Step 7: Retrival(R)
retrived_docs=retriever.get_relevant_documents(query=query)
```

```python
retrived_docs
```

```python
context="".join([doc.page_content for doc in retrived_docs])
context
```

```python
# Step 8: To write a Augmented prompt(A)
prompt=f'''you are a helpful assistant using RAG
here is the context:{context}

The query asked by user is as follows:{query}'''
```

```python
#step 9: Generate the response(G)
print(llm_model.generate_content(prompt).text)
```
